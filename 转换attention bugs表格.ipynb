```python
import os
import re

os.chdir('/Users/rachelwang/monitor-attention')

with open('atqu.txt', 'r') as f:
    c = f.read()

#p = re.compile(r'([\[\]])')
n_c = re.sub('\[', '\t[', c)
new_c = re.sub('\]', ']\t', n_c)

c_lines = new_c.split('\n')

print(c_lines)  
```

```python
import xlsxwriter

workbook =xlsxwriter.Workbook('eg.xlsx')
worksheet = workbook.add_worksheet('sheet1')

worksheet.set_column('A:A', 30)

red = workbook.add_format({'color': 'red'})
blue = workbook.add_format({'color': 'blue'})
bold = workbook.add_format({'bold': True})
italic = workbook.add_format({'italic':True})
center = workbook.add_format({'align':'center'})
superscript = workbook.add_format({'font_script': 1})

ls = ['this is', red, 'red', 'and that is', blue, 'blue']

i=1
for c_line in c_lines:
    chain =c_line.split('\t')
    co = 'B'+str(i)
    if len(chain) == 4:
        worksheet.write_rich_string(co, chain[0], red, chain[1], chain[2], blue, chain[3])
        i +=1
    elif len(chain) ==5:
        worksheet.write_rich_string(co, chain[0], red, chain[1], chain[2], blue, chain[3], chain[4])
        i +=1
    else:
        worksheet.write_rich_string(co, c_line)
        i +=1
```
