```python
import os
os.chdir('/Users/rachelwang/shoe/sizeRevision')
with open('all.tab', 'r') as f1:
    all_lines = f1.readlines()
    
with open('sizeRevision.tsv', 'r') as f2:
    mod_lines = f2.readlines()

all_chains = [all_line.split('\t') for all_line in all_lines]
mod_chains = [mod_line.split('\t') for mod_line in mod_lines]
```

```python

for all_chain in all_chains:
    for mod_chain in mod_chains:
        if all_chain[0] == mod_chain[0] and all_chain[4] == mod_chain[4]:
            all_chain[1] = mod_chain[1]
        
```

```python
new_all_lines = ['\t'.join(all_chain) for all_chain in all_chains]
with open('new.all.tab', 'w') as w3:
    w3.writelines(new_all_lines)
```
